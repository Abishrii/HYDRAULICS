```python
def extract_features(df, sensor_name):

    features = pd.DataFrame()

    features[f"{sensor_name}_mean"] = df.mean(axis=1)
    features[f"{sensor_name}_std"] = df.std(axis=1)
    features[f"{sensor_name}_min"] = df.min(axis=1)
    features[f"{sensor_name}_max"] = df.max(axis=1)

    # RMS = Root Mean Square
    features[f"{sensor_name}_rms"] = np.sqrt((df**2).mean(axis=1))

    # Peak-to-Peak = Maximum - Minimum
    features[f"{sensor_name}_p2p"] = df.max(axis=1) - df.min(axis=1)

    return features
```

```python
healthy_data = master_dataset[
    (master_dataset["Cooler"] == 100) &
    (master_dataset["Valve"] == 100) &
    (master_dataset["Pump"] == 0) &
    (master_dataset["Accumulator"] == 130) &
    (master_dataset["Stable"] == 0)
]

print(healthy_data.shape)
```

```python
fault_data = master_dataset.drop(
    healthy_data.index
)

print(fault_data.shape)
```

```python
# Target columns
label_columns = [
    "Cooler",
    "Valve",
    "Pump",
    "Accumulator",
    "Stable"
]

# Training features
X_train = healthy_data.drop(columns=label_columns)

print(X_train.shape)
```

```python
healthy_data = master_dataset[
    master_dataset["Stable"] == 0
]
```

```python
healthy_data.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = healthy_data.drop(columns=label_columns)

X_train_scaled = scaler.fit_transform(X_train)
```

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

input_dim = X_train_scaled.shape[1]

input_layer = Input(shape=(input_dim,))

encoded = Dense(64, activation="relu")(input_layer)
encoded = Dense(32, activation="relu")(encoded)
encoded = Dense(16, activation="relu")(encoded)

decoded = Dense(32, activation="relu")(encoded)
decoded = Dense(64, activation="relu")(decoded)
decoded = Dense(input_dim, activation="linear")(decoded)

autoencoder = Model(input_layer, decoded)

autoencoder.compile(
    optimizer="adam",
    loss="mse"
)

autoencoder.summary()
```

```python
history = autoencoder.fit(
    X_train_scaled,
    X_train_scaled,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    shuffle=True,
    verbose=1
)
```

```python
reconstructed = autoencoder.predict(X_train_scaled)

reconstruction_error = np.mean(
    np.square(X_train_scaled - reconstructed),
    axis=1
)
```

```python
threshold = np.percentile(
    reconstruction_error,
    95
)

print(threshold)
```

```python
health_score = np.clip(
    100 * (1 - reconstruction_error / threshold),
    0,
    100
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")

plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Autoencoder Training History")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(train_error, bins=40)

plt.xlabel("Reconstruction Error")
plt.ylabel("Frequency")
plt.title("Training Reconstruction Error")

plt.show()
```

```python
X_all = master_dataset.drop(columns=label_columns)

X_all_scaled = scaler.transform(X_all)

all_predictions = autoencoder.predict(X_all_scaled)

all_error = np.mean(
    np.square(X_all_scaled - all_predictions),
    axis=1
)
```

```python
all_error.shape
```

```python
master_dataset["Reconstruction_Error"] = all_error

master_dataset["Anomaly"] = (
    master_dataset["Reconstruction_Error"] > threshold
)
```

```python
master_dataset["Health_Score"] = np.clip(
    100 * (
        1 -
        master_dataset["Reconstruction_Error"] / threshold
    ),
    0,
    100
)
```

```python
master_dataset[[
    "Health_Score",
    "Reconstruction_Error",
    "Anomaly"
]].head(70)
```

```python
feature_errors.shape
```

```python
feature_error_df = pd.DataFrame(
    feature_errors,
    columns=X_all.columns
)

feature_error_df.head()
```

```python
top_features = []

for i in range(len(feature_error_df)):

    errors = feature_error_df.iloc[i]

    top3 = errors.sort_values(
        ascending=False
    ).head(3)

    top_features.append([
        top3.index[0],
        top3.index[1],
        top3.index[2]
    ])
```

```python
top_feature_df = pd.DataFrame(
    top_features,
    columns=[
        "Top_Feature_1",
        "Top_Feature_2",
        "Top_Feature_3"
    ]
)

top_feature_df.head()
```

```python
master_dataset = pd.concat(
    [master_dataset, top_feature_df],
    axis=1
)
```

```python
def health_status(score):

    if score >= 90:
        return "Healthy"

    elif score >= 70:
        return "Monitor"

    elif score >= 40:
        return "Warning"

    else:
        return "Critical"


master_dataset["Health_Status"] = (
    master_dataset["Health_Score"]
    .apply(health_status)
)

master_dataset[["Health_Score","Health_Status"]].head()
```

```python
# Features created during feature engineering
feature_columns = feature_dataset.columns.tolist()

# Create RF input dataset
rf_features = master_dataset[feature_columns].copy()

# Add Autoencoder outputs
rf_features["Health_Score"] = master_dataset["Health_Score"]
rf_features["Reconstruction_Error"] = master_dataset["Reconstruction_Error"]

print(rf_features.shape)
rf_features.head()
```

```python
rf_targets = master_dataset[
    ["Cooler", "Valve", "Pump", "Accumulator"]
].copy()

print(rf_targets.shape)
rf_targets.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    rf_features,
    rf_targets,
    test_size=0.2,
    random_state=42
)

print("X Train :", X_train.shape)
print("X Test  :", X_test.shape)
print("Y Train :", y_train.shape)
print("Y Test  :", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_models = {}

targets = [
    "Cooler",
    "Valve",
    "Pump",
    "Accumulator"
]

for target in targets:

    rf = RandomForestClassifier(
        n_estimators=200,
        max_depth=20,
        random_state=42,
        n_jobs=-1
    )

    rf.fit(
        X_train,
        y_train[target]
    )

    rf_models[target] = rf

print(rf_models.keys())
```

```python
predictions = {}

for target in targets:

    predictions[target] = rf_models[target].predict(
        X_test
    )
```

```python
predictions["Pump"]
```

```python
from sklearn.metrics import accuracy_score

for target in targets:

    acc = accuracy_score(
        y_test[target],
        predictions[target]
    )

    print(f"{target} Accuracy : {acc:.4f}")
```

```python
from sklearn.metrics import classification_report

for target in targets:

    print("="*60)
    print(target.upper())
    print("="*60)

    print(
        classification_report(
            y_test[target],
            predictions[target]
        )
    )
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(
    rf_models["Pump"],
    X_test,
    y_test["Pump"]
)

plt.show()
```
