```python
sensor_files = [
    "PS1","PS2","PS3","PS4","PS5","PS6",
    "FS1","FS2",
    "TS1","TS2","TS3","TS4",
    "VS1",
    "CE","CP","EPS1"
]
```

```python
sensor_data = {}
```

```python
for sensor in sensor_files:

    df = pd.read_csv(
        DATASET_PATH / f"{sensor}.txt",
        sep="\t",
        header=None
    )

    sensor_data[sensor] = df
```

```python
sensor_data.keys()
```

```python
def inspect_sensor(sensor_name, df):

    print("=" * 50)

    print(f"Sensor : {sensor_name}")

    print(f"Shape : {df.shape}")

    print(f"Data Type : {df.dtypes.unique()}")

    print(f"Missing Values : {df.isnull().sum().sum()}")

    print("=" * 50)
```

```python
for sensor_name, df in sensor_data.items():
    inspect_sensor(sensor_name, df)
```
